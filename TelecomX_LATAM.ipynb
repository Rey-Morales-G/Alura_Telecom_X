#Understanding the Churn#
The majority of customers did not churn.
However, a significant portion of the customer base has churned, indicating that churn prediction is a relevant business problem.


```python
import matplotlib.pyplot as plt

# Count churn values
churn_counts = df["Churn"].value_counts()

# Bar chart
plt.figure()
churn_counts.plot(kind="bar")
plt.title("Churn Distribution Among Customers")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.xticks(rotation=0)
plt.show()
```

```python
'Churn vs contract type'
import matplotlib.pyplot as plt

# Create a table: contract type vs churn count
contract_churn = df.groupby(["account.Contract", "Churn"]).size().unstack()

# Plot
plt.figure()
contract_churn.plot(kind="bar")
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=45)
plt.show()
```

```python
'Churn vs Tenure'
plt.figure()
df[df["Churn"] == "Yes"]["customer.tenure"].hist(bins=30)
plt.title("Tenure Distribution â€” Churned Customers")
plt.xlabel("Tenure (months)")
plt.ylabel("Number of Customers")
plt.show()

plt.figure()
df[df["Churn"] == "No"]["customer.tenure"].hist(bins=30)
plt.title("Tenure Distribution â€” Non-Churned Customers")
plt.xlabel("Tenure (months)")
plt.ylabel("Number of Customers")
plt.show()
```

```python
'churn bs monthly charges '
plt.figure()
df[df["Churn"] == "Yes"]["account.Charges.Monthly"].hist(bins=30)
plt.title("Monthly Charges â€” Churned Customers")
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.show()

plt.figure()
df[df["Churn"] == "No"]["account.Charges.Monthly"].hist(bins=30)
plt.title("Monthly Charges â€” Non-Churned Customers")
plt.xlabel("Monthly Charges")
plt.ylabel("Number of Customers")
plt.show()
```
